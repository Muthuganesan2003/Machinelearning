```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

```

```python
dataset=pd.read_csv('covid_19_india (1).csv')
dataset.head()
```

```python
dataset['State/UnionTerritory'].unique()
```

```python
dataset['State/UnionTerritory'].value_counts()
```

```python
covid_Kerala=dataset.loc[(dataset['State/UnionTerritory']=='Kerala')]
covid_Kerala.head(37)
```

```python
plt.plot(covid_Kerala['Date'],covid_Kerala['Cured'],label='Cured Rate')
plt.plot(covid_Kerala['Date'],covid_Kerala['Deaths'],label='Death Rate')
plt.xlabel('Dates')
plt.ylabel('count')
plt.legend()
plt.show()
```

```python
covid_Kerala.isnull().sum()
```

```python
covid_Kerala.replace({"ConfirmedIndianNational":{'-':0},"ConfirmedForeignNational":{'-':0}},inplace=True)
```

```python
x=covid_Kerala.iloc[:,4:-1].values
y=covid_Kerala.iloc[:,-1].values
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor()
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_test)
print(y_pred)
```

```python
print(y_test)
```

```python
np.set_printoptions(precision=0)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
df=pd.DataFrame({'Actual':y_test.flatten(),'Predicted':y_pred.flatten()})
df
```

```python
#Visualization of predicted covid confirmed values in Kerala
df.plot(kind='bar',figsize=(8,8))
plt.title('Visualization of predicted covid confirmed values in Kerala')
plt.xlabel('Number of test values')
plt.ylabel('Number of confirmed cases')
plt.show()
```
