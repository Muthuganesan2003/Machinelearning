```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips .tail()
```

```python
sns.set()
sns.scatterplot(x='tip',y='total_bill',data=tips,hue='day',size='size',palette="YlGnBu")
```

```python
sns.histplot(tips['tip'],kde=True)
```

```python
sns.distplot(tips['tip'],kde=True)
```

```python
sns.displot(tips['tip'],kde=True)
```

```python
sns.barplot(x="sex",y="tip",data=tips)
```

```python
sns.countplot(x='sex',data=tips,hue='sex',palette="YlGnBu")
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue='sex',dodge=True)
```

```python
 sns.jointplot(x="tip",y="total_bill",data=tips,kind="reg")
```

```python
 sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde",shade=True,cmap='YlGnBu')
```
