```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('wine-clustering.csv')
x=dataset.iloc[:,:].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Predicting optimal number of clusters')
plt.xlabel('Number pf clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
kmeans.fit(x)
y_pred=kmeans.predict(x)
print(y_pred)
```

# Parallel coordinates plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data for demonstration
np.random.seed(0)
n_samples = 178
n_features = 13
n_clusters = 4

data = np.random.rand(n_samples, n_features)
labels = np.random.randint(0, n_clusters, n_samples)

# Create a parallel coordinates plot
plt.figure(figsize=(10, 6))

# Plot each data point as a line
for i in range(len(data)):
    plt.plot(range(n_features), data[i], color=plt.cm.Set1(labels[i]), alpha=0.5)

# Customize the plot appearance
plt.xticks(range(n_features), [f"Feature {i}" for i in range(n_features)])
plt.xlabel('Features')
plt.ylabel('Feature Values')
plt.title('Parallel Coordinates Plot')

# Add a legend for the cluster colors
legend_handles = []
for i in range(n_clusters):
    legend_handles.append(plt.Line2D([], [], color=plt.cm.Set1(i), label=f'Cluster {i}'))
plt.legend(handles=legend_handles)

plt.show()
```

# PCA scatter plot


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Generate some sample data for demonstration
np.random.seed(0)
n_samples = 178
n_features = 13
n_clusters = 4

data = np.random.rand(n_samples, n_features)
labels = np.random.randint(0, n_clusters, n_samples)

# Perform PCA for dimensionality reduction
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(data)

# Create a scatter plot of the reduced data
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap=plt.cm.Set1)
plt.colorbar(label='Cluster')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Scatter Plot')

plt.show()
```

#Heat map

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate some sample data for demonstration
np.random.seed(0)
n_features = 13

data = np.random.rand(n_features, n_features)

# Create a heatmap
plt.figure(figsize=(8, 6))

# Plot the heatmap
sns.heatmap(data, cmap='YlGnBu', annot=True, fmt=".2f")

# Customize the plot appearance
plt.xticks(np.arange(n_features) + 0.5, [f"Feature {i}" for i in range(n_features)])
plt.yticks(np.arange(n_features) + 0.5, [f"Feature {i}" for i in range(n_features)])
plt.xlabel('Features')
plt.ylabel('Features')
plt.title('Heatmap')

plt.show()
```
