# Movie Recommendation System using Cosine Similarity

```python
#importing dependencies
import numpy as np
import pandas as pd
```

```python
#reading the dataset
movies_dataset=pd.read_csv('movies.csv')
```

```python
movies_dataset.head()
```

```python
# selecting the relevant features for recommendation

selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
movies_dataset.isnull().sum()
```

```python
# replacing the null valuess with null string

for i in selected_features:
  movies_dataset[i] = movies_dataset[i].fillna('')
```

```python
movies_dataset.isnull().sum()
```

```python
movies_dataset.shape
```

```python
# combining all the 5 selected features

combined_features = movies_dataset['genres']+' '+movies_dataset['keywords']+' '+movies_dataset['tagline']+' '+movies_dataset['cast']+' '+movies_dataset['director']
```

```python
print(combined_features)
```

```python
#converting textual data to numerical data
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
feature_extracted=vectorizer.fit_transform(combined_features)
print(feature_extracted)
```

```python
#finding the similarity score for every single datapoint with all other datapoint
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(feature_extracted)
print(similarity)
```

```python
similarity.shape
```

```python
#reading the movie name
movie=input("Enter the movie name:")
```

```python
#adding all the title value to a list
movie_list=movies_dataset['title'].tolist()
print(movie_list)
```

```python
#Getting a close match of a read movie from user with dataset
import difflib
find_close_match=difflib.get_close_matches(movie,movie_list)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print("The close match is",close_match)
```

```python
#getting index of the movie
movie_index=movies_dataset[movies_dataset.title==close_match]['index'].values[0]
print(movie_index)
```

```python
#Similarity score for a particular movie with all movies
similarity_score=list(enumerate(similarity[movie_index]))
print(similarity_score)
```

```python
#Sorting the similarity scores
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#Printing the top 10 similar movies with the selected movie
i=1
for j in sorted_similar_movies:
  index=j[0]
  recommended_movie=movies_dataset[movies_dataset.index==index]['title'].values[0]
  if(i<=10):
    print(i,') ',recommended_movie,'\n')
    i+=1
```

```python
#movie recommendation system
movie=input("Enter the movie name: ")
print("\n#########################################################################################################################\n")
movie_list=movies_dataset['title'].tolist()
find_close_match=difflib.get_close_matches(movie,movie_list)
close_match=find_close_match[0]
movie_index=movies_dataset[movies_dataset.title==close_match]['index'].values[0]
similarity_score=list(enumerate(similarity[movie_index]))
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
i=1
print("The recommended movies are:\n")
for j in sorted_similar_movies:
  index=j[0]
  recommended_movie=movies_dataset[movies_dataset.index==index]['title'].values[0]
  if(i<=10):
    print(i,') ',recommended_movie,'\n')
    i+=1
print("#########################################################################################################################")
```
